```python
from langchain_community.document_loaders import PyPDFLoader
from langchain_google_genai import GoogleGenerativeAIEmbeddings
from langchain.text_splitter import RecursiveCharacterTextSplitter
from pinecone.grpc import PineconeGRPC as Pinecone
from pinecone import ServerlessSpec
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain
```

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM,pipeline
from langchain.llms import HuggingFacePipeline

tokenizer = AutoTokenizer.from_pretrained("t5-base")
llm_model = AutoModelForSeq2SeqLM.from_pretrained("t5-base")

# Create a pipeline for Hugging Face model
pipe = pipeline("text2text-generation", model=llm_model, tokenizer=tokenizer)

# Use the HuggingFacePipeline from LangChain to wrap the pipeline
llm = HuggingFacePipeline(pipeline=pipe)
```

```python
def load_documents(document_path):
  loader = PyPDFLoader(document_path)
  return loader.load()
```

```python
def text_splitter(documents):
    text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200)

    chunks = []
    for document in documents:
        chunks.extend(text_splitter.split_text(document.page_content))
    return chunks
```

```python
def vector_store(test_chunks):
  vector = []
  for i,chunk in enumerate(test_chunks):
    embedding_vec = model.encode(chunk)
    vector.append((f"doc_{i}",embedding_vec))
  return vector
```

```python
# def get_conversation_chain(query,context):

#   template = """
#         answer the questino {question} based on the given context {context},if u don't know the answer just say that you don't know,dont try to make up an answer.
#   """
#   prompt = PromptTemplate(template=template,input_variables=["question","context"])
#   # chain = LLMChain(prompt=prompt,llm = llm)
#   chain = prompt | llm
#   context = str(context)
#   query = str(query)
#   response = chain.invoke({"question":query,"context":context})
#   return response.replace("\n","")

def get_conversation_chain(query, context):
    template = """
    Answer the question: {question} based on the given context: {context}. If you don't know the answer, say that you don't know. Don't try to make up an answer.
    """
    prompt = PromptTemplate(template=template, input_variables=["question", "context"])
    chain = LLMChain(prompt=prompt, llm=llm)

    # Tokenize query and context
    inputs = tokenizer(query, return_tensors="pt")

    # Run chain
    response = chain.invoke({"question": query, "context": context,"max_new_tokens": 150})
    if "text" in response:
        return response["text"].replace("\n", "")
    else:
        return "No valid response generated."

```

```python
def retrieve_answer(query):
    query_vec = model.encode(query)
    doc_search = index.query(query_vec,top_k=3)
    context = " ".join([doc.page_content for doc in doc_search.matches if doc.page_content is not None])
    response = get_conversation_chain(query,context)
    return response
```

```python
def ask_question(query,pdf_path):
  data = load_documents(pdf_path)
  chunks = text_splitter(data)
  vector = vector_store(chunks)
  index.upsert(vector)
  query_vec = model.encode(query)
  context = retrieve_answer(query)
  response = get_conversation_chain(query,context)
  print(response)

```
